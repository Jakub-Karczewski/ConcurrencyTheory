```python
import numpy as np
import time
from matplotlib import pyplot as plt
import pandas as pd
```

```python
wariant_ = 2
assert wariant_ > 1 and wariant_ <= 6
myPath = "C:\\Users\\HP\\IdeaProjects\\TW-4"
fin = open(myPath + "\\out_" + str(wariant_) + ".txt", "r")
times_str = []
count = 0

for line in fin:
    count += 1
    X = line.strip('\n')
    cut_ = X.split(";")
    times_str.append(cut_)

times_int = [ [] for _ in range(count)]
for i in range(len(times_str)):
    for j in range(len(times_str[i])):
        times_int[i].append(int(times_str[i][j]))

df = []
for X in times_int:
    df.append(np.array(X)/1e6)
    #print(np.array(X)/1e6)

means = [X.mean() for X in df ]
print(means)

```

```python
fig, ax = plt.subplots()
ax.set_title("Åšrednia czasu oczekiwania na swoja kolej dla " + str(count) + " filozofow")
ax.bar(list(range(0, count)), means)
ax.legend()
```

```python
plt.figure(figsize = (12, 12))
Box1 = plt.boxplot(df, showfliers = False)
plt.title("Boxplot dla kazdego z " + str(count) + " filozofow")
plt.ylabel("Czas oczekiwania [ms]")
plt.xlabel("Nr filozofa")
plt.show()
```
